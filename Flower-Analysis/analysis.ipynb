```python
k_rng =range(1,10)
sse =[]
for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit(df[["petal_length", "petal_width"]])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel("k_rng")
plt.ylabel("Sum of squared errors")
plt.plot(k_rng, sse)
```

```python
km = KMeans(n_clusters=3, random_state=0)
y_predict = km.fit_predict(df[["petal_length", "petal_width"]])
y_predict
```

```python
df["cluster"]=y_predict
df.head(100)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df.species, df.cluster)
cm
```

```python
true_fact = df.species
predic_label = df.cluster
cm = confusion_matrix(true_fact, predic_label)
class_label = ["versicolor", "Setosa", "virginica"]

#Plot for confusion matrix
plt.imshow(cm, interpolation="nearest", cmap= plt.cm.Blues)
plt.title("Confusion Matrix")
plt.colorbar()
tick_marks = np.arange(len(class_label))
plt.xticks(tick_marks, class_label)
plt.yticks(tick_marks, class_label)

#Fill matrix with values
for i in range(len(class_label)):
    for j in range(len(class_label)):
        plt.text(j,i, str(cm[i])[j], ha="center", va="center", color="white")

plt.xlabel("predic_label")
plt.ylabel("true_fact")
plt.show()
```
